```python
import pandas as pd
```

```python
df = pd.read_excel("dataset/Pokemon.xlsx")
df.head()
```

# Análisis Descriptivo del Dataset Pokémon

## 1. Dimensiones del Dataset

```python
print(f"Dimensiones del dataset: {df.shape[0]} filas x {df.shape[1]} columnas")
print(f"\nNombres de las columnas:")
print(df.columns.tolist())
```

## 2. Información General del Dataset

```python
print("Tipos de datos en el dataset:")
print(df.dtypes)
print("\n" + "="*50)
print("Valores nulos por columna:")
print(df.isnull().sum())
print("\n" + "="*50)
print("Porcentaje de completitud por columna:")
completitud = (1 - df.isnull().sum() / len(df)) * 100
print(completitud)
```

## 3. Estadísticas Descriptivas

```python
df.describe()
```

## 4. Análisis de Variables Categóricas

```python
print("Columnas de tipo objeto (categóricas):")
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    print(f"\n{col}:")
    print(f"  Valores únicos: {df[col].nunique()}")
    print(f"  Top 5 valores:")
    print(df[col].value_counts().head())
```

## 5. Matriz de Correlación de Variables Numéricas

```python
correlation_matrix = df.select_dtypes(include=['float64', 'int64']).corr()
print("Matriz de Correlación:")
print(correlation_matrix)
print("\nPares de variables más correlacionadas (excluyendo la diagonal):")
# Obtener correlaciones altas
corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_value = correlation_matrix.iloc[i, j]
        if abs(corr_value) > 0.7:
            corr_pairs.append((correlation_matrix.columns[i], correlation_matrix.columns[j], corr_value))

corr_pairs.sort(key=lambda x: abs(x[2]), reverse=True)
for col1, col2, corr in corr_pairs:
    print(f"  {col1} <-> {col2}: {corr:.4f}")
```

## 6. Resumen y Conclusiones

```python
print("=" * 70)
print("RESUMEN DEL ANÁLISIS DESCRIPTIVO")
print("=" * 70)
print(f"\n📊 TAMAÑO DEL DATASET:")
print(f"   • Total de registros: {df.shape[0]}")
print(f"   • Total de atributos: {df.shape[1]}")

print(f"\n🔍 CALIDAD DE DATOS:")
missing_pct = (df.isnull().sum().sum() / (df.shape[0] * df.shape[1])) * 100
print(f"   • Completitud general: {100 - missing_pct:.2f}%")
print(f"   • Filas sin valores nulos: {(df.isnull().sum(axis=1) == 0).sum()} ({((df.isnull().sum(axis=1) == 0).sum() / len(df)) * 100:.1f}%)")

print(f"\n📈 VARIABLES NUMÉRICAS:")
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
print(f"   • Cantidad: {len(numeric_cols)}")
print(f"   • Columns: {', '.join(numeric_cols)}")

print(f"\n🏷️  VARIABLES CATEGÓRICAS:")
object_cols = df.select_dtypes(include='object').columns.tolist()
print(f"   • Cantidad: {len(object_cols)}")
print(f"   • Columns: {', '.join(object_cols)}")

print("\n" + "=" * 70)
```
